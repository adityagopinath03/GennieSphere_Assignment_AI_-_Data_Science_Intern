```python
from google.colab import drive
import pandas as pd

# Mount Google Drive (if not already mounted)
drive.mount('/content/drive')

# Load cleaned dataset
file_path = '/content/drive/MyDrive/DataSet/cleaned_DataScientist.csv'
df = pd.read_csv(file_path)

# Preview location column
df['Location'].head(10)
```

```python
# Step 1: Count frequency of each location (city,state)
top_locations = df['Location'].value_counts().head(10)

# Display results
print("Top 10 Job Locations:\n")
print(top_locations)
```

```python
# Step 2: Visualize Location Insights

import matplotlib.pyplot as plt
import seaborn as sns

# Convert to DataFrame for plotting
top_locations_df = top_locations.reset_index()
top_locations_df.columns = ['Location', 'Count']

# Plot
plt.figure(figsize=(10, 6))
sns.barplot(x='Count', y='Location', data=top_locations_df, palette='plasma')
plt.title('Top 10 U.S. Cities Hiring for Data Roles')
plt.xlabel('Number of Job Postings')
plt.ylabel('City')
plt.tight_layout()
plt.show()
```

## Day 5: Location Insights & Geo Visualization

**Dataset Used:** `cleaned_DataScientist.csv`

---

### Objective:
To identify the most common job locations and visualize geographic hotspots for data science roles.

---

### Steps Performed:

1. **Loaded Cleaned Dataset**  
   - Imported from Google Drive using `pandas`.

2. **Analyzed Job Locations**  
   - Used `Location` column (`city, state` format)
   - Counted the top 10 most frequent job locations

3. **Created `State` Column**  
   - Extracted U.S. state abbreviations (e.g., `ny`, `ca`, `tx`) using regex
   - Used this for state-level grouping

4. **Visualized with Bar Chart**  
   - Highlighted top 10 job hubs including:
     - New York, NY
     - Austin, TX
     - Chicago, IL
     - Houston, TX
     - Los Angeles, CA

5. **Mapped U.S. State-Level Demand with Plotly Choropleth**  
   - Used `plotly.express.choropleth` to visualize posting counts by U.S. state
   - Applied `Viridis` color scale for readability

---

### Key Findings:
- **Top Cities:** New York, Austin, and Chicago dominate the job market  
- **Top States:** Texas, California, New York have the most job openings for data roles  
- Jobs are concentrated in tech-heavy metro areas
